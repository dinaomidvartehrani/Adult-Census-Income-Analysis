# **Supervised learning Classification with Decision Trees**

```python
# Identify the target variable and features
target = 'income'
features = ['age', 'workclass', 'fnlwgt', 'education', 'education.num', 'marital.status',
            'occupation', 'relationship', 'race', 'sex', 'capital.gain', 'capital.loss',
            'hours.per.week', 'native.country']

# Split the dataset into features (X) and target variable (y)
X = balanced_dataset[features]
y = balanced_dataset[target]

#Perform one-hot encoding
X_encoded = pd.get_dummies(X)

# Split the data into training, validation, and test sets (60% training, 20% validation, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)


# Create and train the decision tree classifier
classifier = DecisionTreeClassifier(max_depth = 11)
classifier.fit(X_train, y_train)

# Make predictions on the validation set
y_val_pred = classifier.predict(X_val)

# Calculate the accuracy of the model on the validation set
val_accuracy = accuracy_score(y_val, y_val_pred)
print(f"Validation Accuracy: {val_accuracy}")

# Make predictions on the test set
y_test_pred = classifier.predict(X_test)

# Calculate the accuracy of the model on the test set
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Split the features and the target variable in the labeled dataset
X_labeled = labeled_df.drop(['income_<=50K', 'income_>50K'], axis=1)
y_labeled = labeled_df_before_encode['income']

# Split the features and the target variable in the unlabeled dataset
X_unlabeled = unlabeled_df.drop(['income_<=50K', 'income_>50K'], axis=1)
y_unlabeled = unlabeled_before_encode['income']


# Initialize the labeled data and labels for semi-supervised learning
labeled_data_ss = labeled_df.copy()
labeled_labels_ss = y_labeled.copy()

y_unlabled_df = y_unlabeled.to_frame()

# Set the number of iterations
n_iterations = 4

# Semi-supervised learning loop
for i in range(n_iterations):

    print(f"Iteration {i+1}")

    # Train a decision tree on the labeled data
    tree = DecisionTreeClassifier(max_depth=6)
    tree.fit(X_labeled, labeled_labels_ss)
    # print(tree.classes_)
    # Predict labels for the unlabeled data
    predicted_labels = tree.predict(X_unlabeled)
    column_names = ['income']
    predicted_labels_df = pd.DataFrame(predicted_labels, columns=column_names)


    probabilities = tree.predict_proba(X_unlabeled)
    confidence_threshold = 0.93  # Adjust this threshold as needed
    confident_indices = []
    labels = []
    index = 0
    for sample in probabilities:
      if sample[0] > confidence_threshold:
        labels.append(0)
        confident_indices.append(index)
      if sample[1] > confidence_threshold:
        labels.append(1)
        confident_indices.append(index)
      index += 1
    # print(labels)
    # print(confident_indices)



    # Add the confident predictions to the labeled data
    labeled_data_ss = pd.concat([labeled_data_ss, unlabeled_df.iloc[confident_indices]],ignore_index=True)
    labeled_labels_ss = pd.concat([labeled_labels_ss, y_unlabeled.iloc[confident_indices]],ignore_index=True)

    # Update the encoding for the newly added labeled data only
    # print(X_unlabeled)
    X_labeled = pd.concat([X_labeled, X_unlabeled.iloc[confident_indices]],ignore_index=True)

    # Evaluate the accuracy of the combined dataset
    accuracy = accuracy_score(y_unlabeled, predicted_labels)
    print("Accuracy:", accuracy)

    # Update the unlabeled data by removing the confidently predicted samples
    X_unlabeled = X_unlabeled.drop(X_unlabeled.index[confident_indices]).reset_index(drop=True)
    y_unlabeled = y_unlabeled.drop(y_unlabeled.index[confident_indices]).reset_index(drop=True)
```

```python
def find_features_with_missing_values(dataset):
    # Replace characters with NaN
    replace_chars = ["\n", "\n?\n", "?","\n?"," ?","? "," ? "," ?\n"]
    if any(char in dataset.values for char in replace_chars):
        # Replace characters with NaN using the replace() function from pandas
        dataset.replace(replace_chars, np.nan, inplace=True)
        # Print a message indicating successful replacement
        print("Successfully replaced characters with NaN.")

    # Find features with missing values
    features_with_null = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 0]
    if not features_with_null:
        # Print a message if no missing values are found in any features
        print("No missing values found in any features.")
    else:
        for feature in features_with_null:
            # Calculate the percentage of missing values in the feature
            percentage_missing = round(dataset[feature].isnull().mean() * 100, 2)
            # Print the feature name and percentage of missing values
            print(f"{feature}: {percentage_missing}%")
    # Return the list of features with missing values
    return features_with_null
```

```python
class Classification_AdultCensus(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features=34, out_features=64)
        self.layer_2 = nn.Linear(in_features=64, out_features=128)
        self.layer_3 = nn.Linear(in_features=128, out_features=64)
        self.layer_4 = nn.Linear(in_features=64, out_features=1)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.1)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.dropout(self.relu(self.layer_1(x)))
        x = self.dropout(self.relu(self.layer_2(x)))
        x = self.dropout(self.relu(self.layer_3(x)))
        network = self.layer_4(x)
        return network
class Classification_AdultCensus2(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features=34, out_features=64)
        # self.layer_2 = nn.Linear(in_features=64, out_features=64)
        self.layer_2 = nn.Linear(in_features=64, out_features=128)
        self.layer_3 = nn.Linear(in_features=128, out_features=64)
        self.layer_4 = nn.Linear(in_features=64, out_features=1)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.1)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.dropout(self.relu(self.layer_1(x)))
        x = self.dropout(self.relu(self.layer_2(x)))
        x = self.dropout(self.relu(self.layer_3(x)))
        network = self.layer_4(x)
        return network
```

```python
# Instantiate the model
model_1 = Classification_AdultCensus().to(DEVICE)
model_2 = Classification_AdultCensus2().to(DEVICE)

# Set up the Stochastic Gradient Descent/Adam optimizers and loss function of Binary Cross Entropy
optimizer1 = torch.optim.SGD(model_1.parameters(), lr=0.01)
optimizer2 = torch.optim.Adam(model_1.parameters(), lr=0.01)
criterion = nn.BCEWithLogitsLoss()

torch.manual_seed(42)

# Set the number of epochs
epochs = 10000

# Put data to the target device
X_train, y_train = X_train.to(DEVICE), y_train.to(DEVICE)
X_test, y_test = X_test.to(DEVICE), y_test.to(DEVICE)

# Build the training and evaluation loop
for epoch in range(epochs):
    ### Training
    model_1.train()
    optimizer1.zero_grad()
    model_2.train()
    optimizer2.zero_grad()
    # Forward pass
    y_logits1 = model_1(X_train).squeeze()
    loss1 = criterion(y_logits1, y_train)
    y_logits2= model_2(X_train).squeeze()
    loss2 = criterion(y_logits2, y_train)
    # Backward pass and optimization
    loss1.backward()
    optimizer1.step()
    loss2.backward()
    optimizer2.step()
    ### Testing model 1
    model_1.eval()
    with torch.no_grad():
        test_logits1 = model_1(X_test).squeeze()
        test_loss1 = criterion(test_logits1, y_test)
        test_pred = torch.round(torch.sigmoid(test_logits1))

    if epoch % 2000 == 0:
        print(f"Epoch: {epoch} | Loss: {loss1.item():.4f} | Test Loss: {test_loss1.item():.4f}")
    ### Testing model 2
    model_2.eval()
    with torch.no_grad():
        test_logits2 = model_2(X_test).squeeze()
        test_loss2 = criterion(test_logits2, y_test)
        test_pred2 = torch.round(torch.sigmoid(test_logits2))

    if epoch % 2000 == 0:
        print(f"Epoch: {epoch} | Loss: {loss2.item():.4f} | Test Loss: {test_loss2.item():.4f}")

train_pred = torch.round(torch.sigmoid(model_1(X_train).squeeze()))
test_pred = torch.round(torch.sigmoid(model_1(X_test).squeeze()))
train_score = classification_report(y_train.cpu().detach().numpy(), train_pred.cpu().detach().numpy(), zero_division=1,digits=4,target_names=['<=50K', '>50K'])
test_score = classification_report(y_test.cpu().detach().numpy(), test_pred.cpu().detach().numpy(), zero_division=1,digits=4,target_names=['<=50K', '>50K'])

print(f"\nTrain Classification Report of Model 1:\n{train_score}")
print(f"\nTest Classification Report of Model 1:\n{test_score}")

train_pred2 = torch.round(torch.sigmoid(model_2(X_train).squeeze()))
test_pred2 = torch.round(torch.sigmoid(model_2(X_test).squeeze()))
train_score2 = classification_report(y_train.cpu().detach().numpy(), train_pred2.cpu().detach().numpy(), zero_division=1,digits=4,target_names=['<=50K', '>50K'])
test_score2 = classification_report(y_test.cpu().detach().numpy(), test_pred2.cpu().detach().numpy(), zero_division=1,digits=4,target_names=['<=50K', '>50K'])

print(f"\nTrain Classification Report of Model 2:\n{train_score2}")
print(f"\nTest Classification Report of Model 2:\n{test_score2}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('/content/adult.csv')  # Replace 'adult_census.csv' with the actual filename of your dataset

# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Select features and target variable
features = df.drop('income', axis=1)  # Replace 'income' with the column name of the target variable
target = df['income']

# Impute missing values with the most frequent value
imputer = SimpleImputer(strategy='most_frequent')
features_imputed = pd.DataFrame(imputer.fit_transform(features), columns=features.columns)

# Encode categorical features
categorical_cols = features_imputed.select_dtypes(include='object').columns
label_encoder = LabelEncoder()
features_encoded = features_imputed.copy()
features_encoded[categorical_cols] = features_imputed[categorical_cols].apply(lambda x: label_encoder.fit_transform(x))

# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
features_tsne = tsne.fit_transform(features_encoded)

# Create a DataFrame for the t-SNE results
tsne_df = pd.DataFrame(data=features_tsne, columns=['Dimension 1', 'Dimension 2'])
tsne_df['Income'] = target

# Plot the t-SNE visualization
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Dimension 1', y='Dimension 2', hue='Income', data=tsne_df, palette='Set1')
plt.title('t-SNE Visualization of Adult Census Income Dataset')
plt.show()
```
